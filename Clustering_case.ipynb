#Step 1: Business Understanding


Um e-commerce do Reino Unido contratou os alunos da DNC para agruparem seus clientes com base nas métricas RFM (discutidas abaixo), para então definir estratégias de fidelização.

RFM:

R (Recency): Tempo que o cliente realizou a última compra (em dias)
F (Frequency): Quantidade de compras realizadas pelo cliente
M (Monetary): Valor do ticket médio gasto pelo cliente
O conjunto de dados fornecido contém transações ocorridas entre dezembro de 2010 e dezembro de 2011.

Seu objetivo é clusterizar os clientes com base em suas métricas de RFM.

#Step 4-5: Modeling e Model evaluation


##KMeans

```python
visualizer = KElbowVisualizer(KMeans(), k=10, timings=False)
visualizer.fit(df_rfm_clip_scaled)
visualizer.show()
```

```python
visualizer = SilhouetteVisualizer(KMeans(n_clusters=4))
visualizer.fit(df_rfm_clip_scaled)
visualizer.show()
```

```python
cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results= []
X = df_rfm_clip_scaled.copy()

for k in range (2,11):
  model = KMeans(n_clusters=k, random_state=0)
  labels = model.fit_predict(X)
  cluster_result_dict = {'k': k}
  cluster_result_dict['inertia']= model.inertia_
  for metric in cluster_metrics:
    cluster_result_dict[metric.__name__] = metric(X, labels)
  cluster_metrics_results.append(cluster_result_dict)

pd.DataFrame(cluster_metrics_results).set_index('k').style.background_gradient()
```

```python
KMeans = KMeans(4)
KMeans_labels = KMeans.fit_predict(df_rfm_clip_scaled)
```

```python
px.scatter_3d(
    df_rfm_clip_scaled,
    x='Recency',
    y='Frequency',
    z='Monetary',
    color=KMeans_labels.astype(str),)
```

##Hierarchical Clustering


```python
visualizer = KElbowVisualizer(AgglomerativeClustering(), k=10, timings=False)
visualizer.fit(df_rfm_clip_scaled)
visualizer.show()
```

```python
cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results= []
X = df_rfm_clip_scaled.copy()

for k in range (2,11):
  model = AgglomerativeClustering(n_clusters=k)
  labels = model.fit_predict(X)
  cluster_result_dict = {'k': k}
  for metric in cluster_metrics:
    cluster_result_dict[metric.__name__] = metric(X, labels)
  cluster_metrics_results.append(cluster_result_dict)

pd.DataFrame(cluster_metrics_results).set_index('k').style.background_gradient()
```

```python
AgglomerativeClustering = AgglomerativeClustering(4)
AgglomerativeClustering_labels = AgglomerativeClustering.fit_predict(df_rfm_clip_scaled)
```

```python
px.scatter_3d(
    df_rfm_clip_scaled,
    x='Recency',
    y='Frequency',
    z='Monetary',
    color=AgglomerativeClustering_labels.astype(str),)
```

##Glaussian Mixture


```python
results = []
k_range = range(1, 9)
covariance_types = ['full', 'tied', 'diag', 'spherical']
X = df_rfm_clip_scaled.copy()
for n_components in k_range:
    for covariance_type in covariance_types:
        mclust = GaussianMixture(n_components=n_components,n_init=10,warm_start=True,covariance_type=covariance_type,random_state=0)
        mclust.fit(X)
        results.append({
        'Bic': mclust.bic(X),
        'n_components': n_components,
        'covariance_type': covariance_type,})
results = pd.DataFrame(results)
results.sort_values('Bic').head()
```

```python
sns.lineplot(x='n_components', y='Bic', hue='covariance_type', data=results)
```

```python
mclust = GaussianMixture(n_components=6,n_init=10,covariance_type='diag')
mclust_labels = mclust.fit_predict(df_rfm_clip_scaled)
```

```python
px.scatter_3d(
    df_rfm_clip_scaled,
    x='Recency',
    y='Frequency',
    z='Monetary',
    color=mclust_labels.astype(str),)
```

#Interpretação dos Clusters


```python
centers = pd.DataFrame(KMeans.cluster_centers_, columns=df_rfm_clip_scaled.columns)
centers
```

```python
fig, axes = plt.subplots(nrows=4, figsize=(10, 8),sharex=True)

for i, ax in enumerate(axes):
    center = centers.loc[i,:]
    maxPC = 1.01 * centers.loc[i,:].max()
    colors = ['green' if l > 0  else 'red' for l in center]
    center.plot.bar(ax=ax, color=colors)
    ax.set_ylabel(f'Cluster{i+1}')
    ax.set_ylim(-maxPC, maxPC)
    ax.axhline(color='gray')
    ax.xaxis.set_ticks_position('none')

plt.tight_layout()
```
